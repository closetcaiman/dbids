# Plots for lab 5


```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

from pathlib import Path
import os
```

```python
CWD = Path().cwd()
LAB_DIR = CWD.parent
DATA_DIR = LAB_DIR / "data"
PLOTS_DIR = LAB_DIR / "media"

os.makedirs(PLOTS_DIR, exist_ok=True)
```

## Exercise 3


```python
event_counts = pd.read_csv(Path(DATA_DIR, "event_counts.csv"))

mean_val = event_counts["event_count"].mean()

plt.figure(figsize=(30, 6))
plt.bar(
    event_counts["day"],
    event_counts["event_count"],
    color="skyblue",
    label="Daily Events",
)
plt.axhline(
    mean_val,
    color="red",
    linestyle="dashed",
    linewidth=2,
    label=f"Mean: {mean_val:.2f}",
)
plt.margins(x=0)
plt.title("Daily Event Distribution")
plt.xlabel("Date")
plt.ylabel("Event Count")
plt.legend()
plt.xticks(rotation=90)
plt.tight_layout()
plt.savefig(Path(PLOTS_DIR, "ex3-event-distribution.png"))
plt.show()
```

```python
max_diff = abs(event_counts["event_count"] - mean_val).max()
top_day_idx = abs(event_counts["event_count"] - mean_val).idxmax()
print(f"Max difference from mean: {max_diff}")
print(f"Day: {event_counts.loc[top_day_idx, 'day']}")
print(f"Mean: {mean_val}")
print(f"Event count on that day: {event_counts.loc[top_day_idx, 'event_count']}")
print(f"Percentage difference: {max_diff / mean_val * 100:.2f}%")
```
